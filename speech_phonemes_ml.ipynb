```

import numpy as np
def get_frequencies(x, language):
    f = np.zeros(41)
    for phoneme in x:
        if phoneme == 'AA':
            f[0] += 1
        elif phoneme == 'AE':
            f[1] += 1
        elif phoneme == 'AH':
            f[2] += 1
        elif phoneme == 'AO':
            f[3] += 1
        elif phoneme == 'AW':
            f[4] += 1
        elif phoneme == 'AY':
            f[5] += 1
        elif phoneme == 'B':
            f[6] += 1
        elif phoneme == 'CH':
            f[7] += 1
        elif phoneme == 'D':
            f[8] += 1
        elif phoneme == 'DH':
            f[9] += 1
        elif phoneme == 'EH':
            f[10] += 1
        elif phoneme == 'ER':
            f[11] += 1
        elif phoneme == 'EY':
            f[12] += 1
        elif phoneme == 'F':
            f[13] += 1
        elif phoneme == 'G':
            f[14] += 1
        elif phoneme == 'HH':
            f[15] += 1
        elif phoneme == 'IH':
            f[16] += 1
        elif phoneme == 'IY':
            f[17] += 1
        elif phoneme == 'JH':
            f[18] += 1
        elif phoneme == 'K':
            f[19] += 1
        elif phoneme == 'L':
            f[20] += 1
        elif phoneme == 'M':
            f[21] += 1
        elif phoneme == 'N':
            f[22] += 1
        elif phoneme == 'NG':
            f[23] += 1
        elif phoneme == 'OW':
            f[24] += 1
        elif phoneme == 'OY':
            f[25] += 1
        elif phoneme == 'P':
            f[26] += 1
        elif phoneme == 'R':
            f[27] += 1
        elif phoneme == 'S':
            f[28] += 1
        elif phoneme == 'SH':
            f[29] += 1
        elif phoneme == 'SIL':
            f[30] += 1
        elif phoneme == 'T':
            f[31] += 1
        elif phoneme == 'TH':
            f[32] += 1
        elif phoneme == 'UH':
            f[33] += 1
        elif phoneme == 'UW':
            f[34] += 1
        elif phoneme == 'V':
            f[35] += 1
        elif phoneme == 'W':
            f[36] += 1
        elif phoneme == 'Y':
            f[37] += 1
        elif phoneme == 'Z':
            f[38] += 1
        elif phoneme == 'ZH':
            f[39] += 1

    if language == 'english':
        pass
    elif language == 'spanish':
        f[40] = 1
    elif language == 'german':
        f[40] = 2
    elif language == 'japanese':
        f[40] = 3
    elif language == 'thai':
        f[40] = 4

    return f

def create_matrix_interpreted(phoneme_list, language):
  language_interpreted = np.zeros((1,41))
  for data in phoneme_list:
    x = get_frequencies(data, language)
    language_interpreted = np.vstack([language_interpreted, x[np.newaxis,:]])
  return language_interpreted[1:, :]

spanish_matrix = create_matrix_interpreted(spanish_phonemes, 'spanish')
english_matrix = create_matrix_interpreted(english_phonemes, 'english')
german_matrix = create_matrix_interpreted(german_phonemes, 'german')
japanese_matrix = create_matrix_interpreted(japanese_phonemes, 'japanese')
thai_matrix = create_matrix_interpreted(thai_phonemes, 'thai')

#matrices
print(spanish_matrix)
print(english_matrix)
print(german_matrix)
print(japanese_matrix)
print(thai_matrix)

#matrix lengths
print(spanish_matrix.shape[0])
print(english_matrix.shape[0])
print(german_matrix.shape[0])
print(japanese_matrix.shape[0])
print(thai_matrix.shape[0])
```

```
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
mlp = MLPClassifier()

#Allows us to do 10-fold cross validation
def ten_fold_mlp(values, targets):
  test_size = round(len(targets) / 10)

  left_cut = 0
  right_cut = 0

  train_accuracy = []
  test_accuracy = []
  for i in range(10):
    left_cut = i * test_size

    #makes it so our cut works for edge cases
    if i == 9:
      right_cut = -1
    else:
      right_cut = (i + 1) * test_size

    test_input = values[left_cut:right_cut]
    test_output = targets[left_cut:right_cut]

    input = np.concatenate((values[0:left_cut], values[right_cut + 1:]))
    output = np.concatenate((targets[0:left_cut], targets[right_cut + 1:]))


    #use your own machine learning model
    mlp.fit(input, output)

    train_accuracy.append(mlp.score(input, output))
    test_accuracy.append(mlp.score(test_input, test_output))

  return (train_accuracy, test_accuracy)

def ten_fold_dtc(values, targets):
  test_size = round(len(targets) / 10)

  left_cut = 0
  right_cut = 0

  train_accuracy = []
  test_accuracy = []
  for i in range(10):
    left_cut = i * test_size

    #makes it so our cut works for edge cases
    if i == 9:
      right_cut = -1
    else:
      right_cut = (i + 1) * test_size

    test_input = values[left_cut:right_cut]
    test_output = targets[left_cut:right_cut]

    input = np.concatenate((values[0:left_cut], values[right_cut + 1:]))
    output = np.concatenate((targets[0:left_cut], targets[right_cut + 1:]))

    #use your own machine learning model
    dtc.fit(input, output)

    train_accuracy.append(dtc.score(input, output))
    test_accuracy.append(dtc.score(test_input, test_output))

  return (train_accuracy, test_accuracy)
 
all_language_matrix = np.vstack([english_matrix, spanish_matrix, german_matrix, japanese_matrix])
np.random.shuffle(all_language_matrix)

all_language_matrix[:, :-1] = \
          (all_language_matrix[:, :-1] - np.min(all_language_matrix[:, :-1], 0)) / \
          (np.max(all_language_matrix[:, :-1], 0) - np.min(all_language_matrix[:, :-1], 0)) #normalize

train_accuracy_list_mlp, test_accuracy_list_mlp = ten_fold_mlp(all_language_matrix[:, :-1], all_language_matrix[:, -1])
train_accuracy_list_dtc, test_accuracy_list_dtc = ten_fold_dtc(all_language_matrix[:, :-1], all_language_matrix[:, -1])

avg_train_mlp = np.mean(train_accuracy_list_mlp)
avg_test_mlp = np.mean(test_accuracy_list_mlp)

print("\nMULTI-LAYER PERCEPTRON\n")
print("list of train and test accuracies")
print(train_accuracy_list_mlp, test_accuracy_list_mlp)
print("\naverage train and test accuracy")
print(avg_train_mlp, avg_test_mlp)

avg_train_dtc = np.mean(train_accuracy_list_dtc)
avg_test_dtc = np.mean(test_accuracy_list_dtc)

print("\nDECISION TREE\n")
print("list of train and test accuracies")
print(train_accuracy_list_dtc, test_accuracy_list_dtc)
print("\naverage train and test accuracy ")
print(avg_train_dtc, avg_test_dtc)


```
